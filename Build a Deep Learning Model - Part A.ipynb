The first part is importing the libraries and the data

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
```

First, we scale the data, we can do it using a scikitlearn class: StandardScaler. This basically does the same as substracting the mean and dividing by the standard deviation

```python
# concrete_data_columns = concrete_data.columns

X = concrete_data.drop(['Strength'], axis = 1)# Inputs
y = concrete_data['Strength'] # Targets
```

Let's split the data into train and test. We put aside 30% of the dataset for testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

Let's check if it went well

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

We create the Sequential model now. The model is named 'model'

As the project states, we have to create a model with:

- One hidden layer of 10 nodes, and a ReLU activation function
- Use the adam optimizer and the mean squared error as the loss function.

```python
from keras.models import Sequential
from keras.layers import Dense
keras.backend.clear_session() # We delete any possible previous session.
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=X_train.shape[1:]))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

We can use the summary() function to see the layers of our model

```python
model = regression_model()
model.summary()
```

Train the model on the training data using 50 epochs.

```python
model.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_test)
mean_squared_error(y_test,y_pred)
```

Looks Good!
Let's create a loop for storing 50 MSE values so that we can calculate the mean MSE

```python
mse_list = []
for i in range(50):
    print('Loop {} out of 50'.format(i+1))
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
    
    keras.backend.clear_session() # We delete any possible previous session.
    model = regression_model()
    model.fit(X_train, y_train, epochs=50, verbose = 0) # No verbosity, so we can see the number of loop
    
    y_pred = model.predict(X_test)
    mse_list.append(mean_squared_error(y_test, y_pred)) 
    
    del model # In order not to store possible weight values, I am deleting the model, so next time it starts over
```

```python
mse_mean = np.mean(mse_list)
mse_std = np.std(mse_list)
```

```python
print("The mean value of the MSE is {:.2f}, and the standard deviation is {:.2f}".format(mse_mean, mse_std))
```
